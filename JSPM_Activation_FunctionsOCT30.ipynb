```python
# importing the libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5,5,0.25)
```

```python
x
```

```python
y = x * 4.67 + 21.34
```

```python
y
```

```python
plt.plot(x,y);
```

```python
def sigmoid(z):
    val = 1/(1+np.exp(-z))
    return val
```

```python
y = sigmoid(x)
```

```python
print(y)
```

```python
plt.plot(x,y)
```

```python
sigmoid(-3)
```

```python
sigmoid(5)
```

```python
sigmoid(0.54)
```

```python
def relu(z):
    if z <= 0:
        return 0
    else:
        return z
```

```python
relx = [relu(a) for a in x]
```

```python
plt.plot(x,relx)
```
